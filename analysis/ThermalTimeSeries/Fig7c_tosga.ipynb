```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
#import seaborn as sns
import pandas as pd
%matplotlib inline
import mplsettings
import glob

from datetime import timedelta

print("numpy version  : ", np.__version__)
print("pandas version : ", pd.__version__)
print("xarray version : ", xr.__version__)
```

```python
#dirstix = "/net2/rnd/dev/stixfonts/fonts/static_ttf"
dirstix = "../../common/static_ttf"
import matplotlib.font_manager as font_manager
for font in font_manager.findSystemFonts(dirstix):
    font_manager.fontManager.addfont(font)

# Set font family globally
plt.rcParams['font.family'] = 'STIX Two Text'
```

```python
dataset4p0  = xr.open_mfdataset('tosga_CM4p0.nc')
datasetp125 = xr.open_mfdataset('tosga_CM4Xp125.nc')
datasetp25  = xr.open_mfdataset('tosga_CM4Xp25.nc')
```

```python
tosga_4p0  = dataset4p0['tosga_4p0']
tosga_p125 = datasetp125['tosga_p125']
tosga_p25  = datasetp25['tosga_p25']
```

```python
fig, axis = plt.subplots(ncols=1, nrows=1, figsize=(6.5,4.0))

tosga_p125.plot.line(ax=axis, label='CM4X-p125',linestyle='solid', color='blue', linewidth=1) 
tosga_p25.plot.line(ax=axis, label='CM4X-p25',linestyle='solid', color='green', linewidth=1) 
tosga_4p0.plot.line(ax=axis, label='CM4.0',linestyle='solid', color='red', linewidth=1) 

axis.legend(loc='upper right')
axis.set_ylim([16.95,17.7])
axis.set_xlim(0,900)
#axis.set_xlim(1750,2400)
axis.set_ylabel(r'$^\circ$C',fontsize=14, labelpad=1)
#axis.set_ylabel('degree C')
axis.set_xlabel('year',fontsize=14)
axis.grid(linewidth=.5)
plt.xticks(ticks=[0, 50, 150, 250, 350, 450, 550, 650, 750, 850, 950, 1050])

axis.set_title(r'Ocean area ave sea surface temperature ($^\circ$C)')
#axis.set_title('Ocean area ave annual sea surface temperature (deg C)')
fig.savefig('Fig07_tosga_piControl.jpg', bbox_inches='tight', dpi=400, format='jpeg')
```
